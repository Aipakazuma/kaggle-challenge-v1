```python
from sklearn.cluster import KMeans
```

```python
features = trainingData.ix [:, ['loc_x', 'loc_y', 'shot_made_flag']].as_matrix()
```

```python
kmeans_model = KMeans(n_clusters=6, random_state=10).fit(features)
```

```python
labels = kmeans_model.labels_
```

```python
pl.pyplot.figure(figsize=(15, 25), dpi=100)
pl.pyplot.scatter(features[0], features[1], c=labels, alpha=0.5)
```
